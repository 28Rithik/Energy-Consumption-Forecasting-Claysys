```python
import pandas as pd

bareilly = pd.read_csv("../data/processed/bareilly_all.csv")
mathura = pd.read_csv("../data/processed/mathura_all.csv")
```

```python
master_df = pd.concat(
    [bareilly, mathura],
    ignore_index=True
)
```

```python
master_df.to_csv(
    "../data/processed/energy_master.csv",
    index=False
)
```
