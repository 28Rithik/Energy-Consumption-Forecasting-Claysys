```python
import numpy as np
import pandas as pd
```

```python
master_df = pd.read_csv("../data/processed/energy_master.csv")
```

```python
master_df["x_Timestamp"] = pd.to_datetime(master_df["x_Timestamp"])

hourly_df = (
    master_df
    .groupby(["city", "meter"])
    .resample("H", on="x_Timestamp")["t_kWh"]
    .sum()
    .reset_index()
)
```

```python
hourly_df.shape
hourly_df.head()
```

```python
hourly_df.to_csv(
    "../data/processed/hourly_energy.csv",
    index=False
)
```

```python
hourly_df["meter"].value_counts().head()
```

```python
meter_df = hourly_df[
    hourly_df["meter"] == "BR02"
].copy()
```

```python
daily_df = (
    meter_df
    .set_index("x_Timestamp")
    .resample("D")["t_kWh"]
    .sum()
    .reset_index()
)
```
