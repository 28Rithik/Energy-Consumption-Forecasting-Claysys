```python
import numpy as np
import pandas as pd
```

```python
master_df = pd.read_csv("../data/processed/energy_master.csv")
```

```python
master_df["x_Timestamp"] = pd.to_datetime(master_df["x_Timestamp"])

hourly_df = (
    master_df
    .groupby(["city", "meter"])
    .resample("H", on="x_Timestamp")["t_kWh"]
    .sum()
    .reset_index()
)
```
