```python
import pandas as pd

hourly_df = pd.read_csv(
    "../data/processed/hourly_energy.csv"
)

hourly_df["x_Timestamp"] = pd.to_datetime(
    hourly_df["x_Timestamp"]
)
```

```python
grid_df = (
    hourly_df
    .groupby("x_Timestamp")
    ["t_kWh"]
    .sum()
    .reset_index()
)

grid_df.head()
```

```python
grid_df.shape
```

```python
grid_df["hour"] = grid_df["x_Timestamp"].dt.hour
grid_df["dayofweek"] = grid_df["x_Timestamp"].dt.dayofweek
grid_df["month"] = grid_df["x_Timestamp"].dt.month
```

```python
grid_df["lag_1"] = grid_df["t_kWh"].shift(1)
grid_df["lag_24"] = grid_df["t_kWh"].shift(24)
grid_df["lag_168"] = grid_df["t_kWh"].shift(168)
```

```python
grid_df["rolling_mean_24"] = (
    grid_df["t_kWh"]
    .rolling(24)
    .mean()
)

grid_df["rolling_std_24"] = (
    grid_df["t_kWh"]
    .rolling(24)
    .std()
)
```

```python
grid_df = grid_df.dropna()
```

```python
grid_df.to_csv(
    "../data/processed/grid_energy.csv",
    index=False
)
```
