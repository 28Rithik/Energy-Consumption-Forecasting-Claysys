```python
import pandas as pd
```

```python
bareilly_files = [
    "../data/raw/Bareilly_2019.csv",
    "../data/raw/Bareilly_2020.csv",
    "../data/raw/Bareilly_2021.csv"
]
```

```python
dfs = []

for file in bareilly_files:
    df = pd.read_csv(file)
    dfs.append(df)
```

```python
bareilly_df = pd.concat(dfs, ignore_index=True)
```

```python
print(bareilly_df.shape)
```

```python
bareilly_df["x_Timestamp"] = pd.to_datetime(
    bareilly_df["x_Timestamp"]
)
```

```python
bareilly_df = bareilly_df.sort_values(
    by="x_Timestamp"
)
```

```python
bareilly_df["city"] = "Bareilly"
```

```python
bareilly_df.to_csv(
    "../data/processed/bareilly_all.csv",
    index=False
)
```

```python
print(bareilly_df.head())
print(bareilly_df.shape)
print(bareilly_df["x_Timestamp"].min())
print(bareilly_df["x_Timestamp"].max())
```
